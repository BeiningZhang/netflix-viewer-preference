```python
sns.set(style="darkgrid")
ax = sns.countplot(x="type", data=df, palette="Set2")
```

It indicates that Netflix has more movies than shows.

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(x="age_certification", data=df, palette="Set2", 
                   order=df['age_certification'].value_counts().index[0:15])
```

* "TV-MA" is a rating assigned by the TV Parental Guidelines to a television program that was designed for mature audiences only.
* "TV-14" stands for content that may be inappropriate for children younger than 14 years of age.
* "R" represents films that has been assessed as having material which may be unsuitable for children under the age of 17 by the Motion Picture Association of America; the MPAA writes "Under 17 requires accompanying parent or adult guardian".

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y="release_year", data=df, palette="Set2", 
                   order=df['release_year'].value_counts().index[0:15])
```
